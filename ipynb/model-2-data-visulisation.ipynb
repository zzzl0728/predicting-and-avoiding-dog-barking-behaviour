```python
import wave
import numpy as np
import matplotlib.pyplot as plt

wav_file = "C:/Users/zzzl0/Desktop/predicting-and-avoiding-dog-barking-behaviour/predicting-and-avoiding-dog-barking-behaviour/output/model2_audio_clips/Dog3_8.wav"
with wave.open(wav_file, 'rb') as wav:
    frames = wav.readframes(-1)
    signal = np.frombuffer(frames, dtype='int16')

    sample_rate = wav.getframerate()
    duration = len(signal) / sample_rate

time = np.linspace(0., duration, len(signal))

plt.figure(figsize=(10, 4))
plt.plot(time, signal)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform of sample audio")
plt.grid(True)
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

wav_file = "C:/Users/zzzl0/Desktop/predicting-and-avoiding-dog-barking-behaviour/predicting-and-avoiding-dog-barking-behaviour/output/model2_audio_clips/Dog3_8.wav"

signal, sample_rate = librosa.load(wav_file)

hop_length = 512
n_fft = 2048
stft = librosa.stft(signal, n_fft=n_fft, hop_length=hop_length)

power_spectrogram = np.abs(stft)**2

plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(power_spectrogram, ref=np.max), sr=sample_rate, hop_length=hop_length, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram of {}'.format(wav_file))
plt.show()
```
